```python
import json, requests

resp = requests.get('https://servicodados.ibge.gov.br/api/v2/cnae/subclasses', timeout=60)
resp.raise_for_status()
data = resp.json()

d = {}
for item in data:
    c = item['id']
    desc = item['descricao'].strip().title()
    fmt = f'{c[:4]}-{c[4]}/{c[5:]}' if len(c) == 7 else c
    d[f'{desc} ({fmt})'] = int(c)
d = dict(sorted(d.items()))

output_path = '/workspaces/pesquisa-preco/cnaes_completo.json'
with open(output_path, 'w', encoding='utf-8') as f:
    json.dump(d, f, ensure_ascii=False, indent=2)

print(f'OK: {len(d)} CNAEs gerados e salvos em {output_path}')
```
